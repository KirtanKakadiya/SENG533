```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_csv = pd.read_csv('CSV-Testing-Results.csv')
df_parquet = pd.read_csv('Parquet-Testing-Results.csv')
```

```python
plt.plot(df_csv['Rows'], df_csv['Mean CSV Busy Time (s)'], marker='o', label='Mean CSV Busy Time (s)')
plt.plot(df_csv['Rows'], df_parquet['Mean Parquet Busy Time (s)'], marker='o', label='Mean Parquet Busy Time (s)')
plt.xlabel('Rows (million)')
plt.ylabel('Mean Busy Time (s)')
plt.title('CSV vs Parquet Busy Time')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.plot(df_csv['Rows'], df_csv['Mean CSV CPU Util (%)'], marker='o', label='Mean CSV CPU Util (%)')
plt.plot(df_csv['Rows'], df_parquet['Mean Parquet CPU Util (%)'], marker='o', label='Mean Parquet CPU Util (%)')
plt.xlabel('Rows (million)')
plt.ylabel('Mean CPU Utilization (%)')
plt.title('CPU Utilization vs Data Size (4 vCPU, 16 GiB)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.plot(df_csv['Rows'], df_csv['Mean CSV Mem Util (%)'], marker='o', label='Mean CSV Mem Util (%)')
plt.plot(df_csv['Rows'], df_parquet['Mean Parquet Mem Util (%)'], marker='o', label='Mean Parquet Mem Util (%)')
plt.xlabel('Rows (million)')
plt.ylabel('Mean Memory Usage (%)')
plt.title('Memory Utilization vs Data Size (4 vCPU, 16 GiB)')
plt.legend()
plt.grid(True)
plt.show()
```
